2. Menghapus beberapa kolom yang tidak akan digunakan

Berdasarkan pertanyaan bisnis yang sudah ada, maka untuk kolom yang akan didrop, yakni:
- instant (Nomor instan): Karena indeks catatan ini tidak memberikan informasi tambahan yang berguna.
- dteday (Tanggal): Tidak melibatkan aspek temporal pada tingkat tanggal.
- casual (Peminjam sepeda kasual): Fokus hanya pada total (cnt), bukan perbedaan antara peminjam kasual dan terdaftar.
- registered (Peminjam sepeda terdaftar): Fokus hanya pada total (cnt), bukan perbedaan antara peminjam kasual dan terdaftar.

```python
drop_col = ['instant', 'dteday', 'casual', 'registered']

for i in day_df.columns:
  if i in drop_col:
    day_df.drop(labels=i, axis=1, inplace=True)

day_df.head()
```

3. Mengubah beberapa detail tentang kolom

Karena nama-nama dari kolom diatas agak kurang sesuai dengan style saya, maka akan saya rubah. Dan juga saya lebih suka dengan tipe data yang bentuknya kategorikal, jadi beberapa yang sebelumnya adalah angka akan saya rubah. Serta sebelumnya ada tipe data yang juga tidak sesuai, dan itu akan saya rubah juga.

```python
# Mengubah nama judul kolom
day_df.rename(columns={
    'yr': 'year',
    'mnth': 'month',
    'weathersit': 'weather_cond',
    'cnt': 'count'
}, inplace=True)

day_df.head()
```

```python
# Mengubah angka menjadi keterangan
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

```python
# Mengubaha tipe data ke categorical
day_df['season'] = day_df.season.astype('category')
day_df['year'] = day_df.year.astype('category')
day_df['month'] = day_df.month.astype('category')
day_df['holiday'] = day_df.holiday.astype('category')
day_df['weekday'] = day_df.weekday.astype('category')
day_df['workingday'] = day_df.workingday.astype('category')
day_df['weather_cond'] = day_df.weather_cond.astype('category')

day_df.head()
```

## Visualization & Explanatory Analysis

6. Kelompokkan penyewa sepeda berdasarkan season

### Pertanyaan 1: Bagaimana pemanfaatan sepeda berubah seiring perubahan kondisi cuaca (weathersit), dan bagaimana bisnis sepeda dapat memanfaatkan informasi ini untuk meningkatkan penawaran atau layanan?

```python
plt.figure(figsize=(10, 6))
sns.barplot(
    x='weather_cond',
    y='count',
    data=day_df,
    palette=['skyblue', 'lightgreen', 'lightcoral']
    )

plt.title('Pemanfaatan Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pemanfaatan Sepeda')
plt.ylim(0, max(day_df['count']) + 500) 

plt.show()
```

Berdasarkan gambaran data yang disajikan melalui boxplot, dapat disimpulkan bahwa ada korelasi yang signifikan antara kondisi cuaca dan jumlah penyewa sepeda. Kondisi cuaca yang cerah atau setengah mendung memunculkan minat yang paling tinggi, diikuti oleh kondisi berkabut atau mendung, sementara kondisi sedikit bersalju atau hujan menunjukkan minat yang paling rendah.

### Pertanyaan 2: Dapatkah analisis pola peminjaman sepeda pada hari kerja (workingday) dan hari libur (holiday) membantu bisnis dalam mengoptimalkan stok sepeda atau menyesuaikan strategi pemasaran untuk meningkatkan penggunaan sepeda?

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(15,10))

# Berdasarkan workingday
sns.barplot(
    x='workingday',
    y='count',
    data=day_df,
    ax=axes[0])
axes[0].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Kerja')
axes[0].set_xlabel('Hari Kerja')
axes[0].set_ylabel('Jumlah Pengguna Sepeda')

# Berdasarkan holiday
sns.barplot(
  x='holiday',
  y='count',
  data=day_df,
  ax=axes[1])
axes[1].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Libur')
axes[1].set_xlabel('Hari Libur')
axes[1].set_ylabel('Jumlah Pengguna Sepeda')

# Berdasarkan weekday
sns.barplot(
  x='weekday',
  y='count',
  data=day_df,
  ax=axes[2])
axes[2].set_title('Jumlah Pengguna Sepeda berdasarkan Hari dalam Seminggu')
axes[2].set_xlabel('Hari dalam Seminggu')
axes[2].set_ylabel('Jumlah Pengguna Sepeda')

plt.tight_layout()
plt.show()
```

Berdasarkan representasi visual data menggunakan barplot di atas, dapat diungkapkan beberapa temuan menarik:
- Terdapat kecenderungan bahwa jumlah peminjam sepeda lebih tinggi pada hari kerja (Senin-Jumat) dibandingkan dengan akhir pekan (Sabtu dan Minggu), di mana 0 mengindikasikan akhir pekan dan 1 mengindikasikan hari kerja.
- Pada umumnya, peminjaman sepeda lebih tinggi saat tidak ada hari libur nasional dibandingkan dengan hari-hari libur tersebut.
- Menariknya, Jumat menjadi hari dengan jumlah peminjam sepeda tertinggi, sementara Minggu menempati posisi terakhir dengan jumlah peminjam sepeda yang paling sedikit.
